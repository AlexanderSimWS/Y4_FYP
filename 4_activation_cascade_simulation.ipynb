# Activation Cascade Simulation
To highlight the difference between a simulation without resource coupling and with resource coupling, an activation cascade from the paper: [Resource Competition Shapes the Response of Genetic Circuits](https://pubs.acs.org/doi/10.1021/acssynbio.6b00361) by Qian. *et al* is used. 

## Activation Cascade Description
- A two-stage activation cascade was built by connecting three nodes in a cascade topology.
- The first node produces the TF LuxR which is activated by AHL.
- The second node is activated by active LuxR and produces NahR which is activated by SAL.
- The third node is activated by activate NahR and produces GFP

## Activation Cascade with Resource Convservation
To simulate the model with resource conservation, we need to calculate the resource demand coefficients ($J = \sum_j k^{m}_{j} \cdot A_j$) and subsequently calculated $k_{eff}$.

We now define the new rate (derived with QSS approximation), $k_{eff;i}$, as

\begin{align*}
    k_{eff;i} = \frac{k_{i}^{\theta} \cdot k^{m}_{i} \cdot R_T}{{1 + \sum_j k^{m}_{j} \cdot A_j}}
\end{align*}

The reaction equations are:

#### Node 1

\begin{align*}
    \varnothing &\xrightarrow{k_{eff,1;1} \cdot g^T_1} m_1 \\
    m_1 &\xrightarrow{k_{eff,2;1}} x^0_1 \\
    m_1 &\xrightarrow{\delta_1} \varnothing \\
    x^0_1 &\xrightarrow{\lambda} \varnothing
\end{align*}

#### Edge 1

\begin{align*}
    x^0_1 + n'_1 \cdot I_1 &\xleftrightharpoons[\kappa_1^-]{\kappa_1^+} x^+_1 \\
    x^+_1 &\xrightarrow{\lambda} \varnothing
\end{align*}

#### Node 2

\begin{align*}
    \varnothing &\xrightarrow{k_{eff,1;2} \cdot g^T_2 \cdot \alpha(x_1, n_1)} m_2 \\
    m_2 &\xrightarrow{k_{eff,2;2}} x^0_2 \\
    m_2 &\xrightarrow{\delta_2} \varnothing \\
    x^0_2 &\xrightarrow{\lambda} \varnothing
\end{align*}

#### Edge 2

\begin{align*}
    x^0_2 + n'_2 \cdot I_2 &\xleftrightharpoons[\kappa_2^-]{\kappa_2^+} x^+_2 \\
    x^+_2 &\xrightarrow{\lambda} \varnothing
\end{align*}

#### Node 3

\begin{align*}
    \varnothing &\xrightarrow{k_{eff,1;3} \cdot g^T_3 \cdot \alpha(x_2, n_2)} m_3 \\
    m_3 &\xrightarrow{k_{eff,2;3}} x_3 \\
    m_3 &\xrightarrow{\delta_3} \varnothing \\
    x_3 &\xrightarrow{\lambda} \varnothing
\end{align*}

Where the activation Hill function is:
\begin{align*}
    \alpha(x^+_j, n_j) = \frac{\hat{k}^m_{a;i} \cdot (x^+_j)^{n_j}}{1 + \hat{k}^m_{a;i} \cdot (x^+_j)^{n_j}}
\end{align*}

```julia
include("DSL_backend/ode_system.jl")

# Units = uM
tx1 = [100,5,95] # km = 1
tx2 = [17,2,100] # km = 1/6
tx3 = [100,5,95] # km = 1
tl1 = [7,1,104] # km = 1/15
tl2 = [7,1,104] # km = 1/15
tl3 = [7,1,104] # km = 1/15

act1 = [10,1] # k =  10 uM
act2 = [100,1] # k = 100 uM
# act3 = [10,1] # k = 10 uM

ind1 = [1,252] # k_hat = 1/252
ind2 = [1,1000] # k_hat = 1/1000

dg = 0.0833 # degradation rate δ

vect = []
conc = [0.01, 0.02, 0.05, 0.08, 0.1, 0.5,1, 5, 10] # AHL concentrations

for i=1:length(conc)
node_1 = createnode(:A, tx1, tl1, dg, 0.06) # Gene concentration = 0.06uM 
link_1 = tfactivation(:I1, node_1, ind1, 1.4, conc[i])

node_2 = createnode(:B, tx2, tl2, dg, 0.06)
link_2 = activate(node_1, node_2, act1, 1) # cooperativity = 1
link_3 = tfactivation(:I2, node_2, ind2, 4, 100) # SAL conc. = 100uM

node_3 = createnode(:C, tx3, tl3, dg, 0.06)
link_4 = activate(node_2, node_3, act2, 1)

compiled_links = compilelinks(link_1, link_2, link_3, link_4)
compiled_nodes = compilenodes(node_1, node_2, node_3)

node_sys = createnodesystem(compiled_nodes, compiled_links)

res = runsystem(node_sys, 1, 1, 0.0067, (0,150)) # ribosome = rnap concentration = 1, λ = 0.0067

push!(vect, maximum(res[10,:]))
end
plot(conc, vect,
    xlabel = "AHL Concentration (uM)",
    ylabel = "GFP Concentration (uM)",
    label = "GFP", fmt = :png, title = "Dose Response With Resource Conservation"
)
# savefig("resource-conservation-simulation")
```

## Simulation of Activation Cascade without Resource Conservation:
To simulate the model without resource conservation, we simply need to remove the resource demand coefficient ($J = \sum_j k^{m}_{j} \cdot A_j$) from $k_{\text{eff}} = \frac{k_{i}^{\theta} \cdot k^{m}_{i} \cdot R_T}{{1 + \sum_j k^{m}_{j} \cdot A_j}}$.

We now define the new rate (derived with QSS approximation), $k_{old;i}$, as

\begin{align*}
    k_{old;i} = k_{i}^{\theta} \cdot k^{m}_{i} \cdot R_T
\end{align*}

The reaction equations are now:

#### Node 1

\begin{align*}
    \varnothing &\xrightarrow{k_{old,1;1} \cdot g^T_1} m_1 \\
    m_1 &\xrightarrow{k_{old,2;1}} x^0_1 \\
    m_1 &\xrightarrow{\delta_1} \varnothing \\
    x^0_1 &\xrightarrow{\lambda} \varnothing
\end{align*}

#### Edge 1

\begin{align*}
    x^0_1 + n'_1 \cdot I_1 &\xleftrightharpoons[\kappa_1^-]{\kappa_1^+} x^+_1 \\
    x^+_1 &\xrightarrow{\lambda} \varnothing
\end{align*}

#### Node 2

\begin{align*}
    \varnothing &\xrightarrow{k_{old,1;2} \cdot g^T_2 \cdot \alpha(x_1, n_1)} m_2 \\
    m_2 &\xrightarrow{k_{old,2;2}} x^0_2 \\
    m_2 &\xrightarrow{\delta_2} \varnothing \\
    x^0_2 &\xrightarrow{\lambda} \varnothing
\end{align*}

#### Edge 2

\begin{align*}
    x^0_2 + n'_2 \cdot I_2 &\xleftrightharpoons[\kappa_2^-]{\kappa_2^+} x^+_2 \\
    x^+_2 &\xrightarrow{\lambda} \varnothing
\end{align*}

#### Node 3

\begin{align*}
    \varnothing &\xrightarrow{k_{old,1;3} \cdot g^T_3 \cdot \alpha(x_2, n_2)} m_3 \\
    m_3 &\xrightarrow{k_{old,2;3}} x_3 \\
    m_3 &\xrightarrow{\delta_3} \varnothing \\
    x_3 &\xrightarrow{\lambda} \varnothing
\end{align*}

Where the activation Hill function is:
\begin{align*}
    \alpha(x^+_j, n_j) = \frac{\hat{k}^m_{a;i} \cdot (x^+_j)^{n_j}}{1 + \hat{k}^m_{a;i} \cdot (x^+_j)^{n_j}}
\end{align*}

```julia
# Import Catalyst.jl for good measure
using Catalyst, Plots, DifferentialEquations

# Functions
k_old(kθ, km, R_T) = kθ * km * R_T # k_old
α(km, x, n) = (km * x^n)/(1 + (km * x^n)) # activation Hill function

# Parameters
p_T = 1 # Total RNAP concentration
r_T = 1 # Total ribosome concentration
k1θ₁ = k1θ₂ = k1θ₃ =  1.67 # Transcrption elongation rate
k2θ₁ = k2θ₂ = k2θ₃ =  1.67 # Translation elongation rate
gᵀ₁ = gᵀ₂ = gᵀ₃ = 0.06 # Total gene concentration = 60 copies
δ₁ = δ₂ = δ₃ = 0.0833 # mRNA degradation rate
λ = 0.0067 # Dilution rate
k1m₁ = k1m₃ = 1 # Binding constant between DNA and RNAP
k1m₂ = 1/6 # Binding constant between DNA and RNAP
k2m₁ = k2m₂ = k2m₃ = 1/15 # Binding constant between mRNA and ribosome
kma₁ = 10 # Binding constant between activated LuxR and DNA 2
kma₂ = 100 # Binding constant between activated NahR and DNA 3
k_hat₁ = [252, 1] # Binding constant between inactive LuxR and AHL 
k_hat₂ = [100, 1] # Binding constant between inactive NahR and SAL
n = 1 # Cooperativity coefficient for activated TFs and DNA
nᴵ1 = 1.4 # Cooperativity coefficient for AHL
nᴵ2 = 4 # Cooperativity coefficient for SAL

AHL = [0.01, 0.02, 0.05, 0.08, 0.1, 0.5,1, 5, 10] # Inducer 1 concentration
SAL = 100 # Inducer 2 concentration

vect= []

# Reaction system specification
@variables t
@species x₁(t) x₂(t)

# Node 1 rates
k1₁ = k_old(k1θ₁, k1m₁, p_T) * gᵀ₁
k2₁ = k_old(k2θ₁, k1m₂, r_T) 

# Node 2 rates
k1₂ = k_old(k1θ₂, k1m₂, p_T) * gᵀ₂ * α(k1m₁, x₁, n)
k2₂ = k_old(k2θ₂, k2m₂, r_T)

# Node 3 rates
k1₃ = k_old(k1θ₃, k1m₃, p_T) * gᵀ₃ * α(k1m₂ , x₂ , n)
k2₃ = k_old(k2θ₃, k2m₂, r_T)

# Find the estimate of the steady-state value for GFP production for all AHL concentrations
for i = 1:length(AHL)
    # Node 1
    node_1 = @reaction_network begin
        @species m₁(t)=0.0 m₂(t)=0.0 m₃(t)=0.0 x₁(t)=0.0 x₂(t)=0.0 x₃(t)=0.0 x⁰₁(t)=0.0 x⁰₂(t)=0.0 I₁(t)=$AHL[$i] I₂(t)=100.0
        $k1₁, ∅ --> m₁
        $k2₁, m₁ --> x⁰₁
        $δ₁, m₁ --> ∅
        $λ, x⁰₁ --> ∅
    end

    # Edge 1
    edge_1 = @reaction_network begin
        ($k_hat₁[1], $k_hat₁[2]), x⁰₁ + $nᴵ1*I₁ <--> x₁
        $λ, x₁ --> ∅
    end

    # Node 2 
    node_2 = @reaction_network begin
        $k1₂, ∅ --> m₂
        $k2₂, m₂ --> x⁰₂
        $δ₂, m₂ --> ∅
        $λ, x⁰₂ --> ∅ 
    end

    # Edge 2
    edge_2 = @reaction_network begin
        ($k_hat₂[1], $k_hat₂[2]), x⁰₂ + $nᴵ2*I₂ <--> x₂
        $λ, x₂ --> ∅
    end

    # Node 3
    node_3 = @reaction_network begin
        $k1₃ , ∅ --> m₃
        $k2₃, m₃ --> x₃
        $δ₃, m₃ --> ∅
        $λ, x₃ --> ∅ 
    end

    # Combine nodes
    @named rxn_sys = extend(extend(extend(extend(node_1, edge_1), node_2), edge_2), node_3)
    # Convert to ODE system
    tspan = (0, 1000)
    u0 = []
    p = []
    ode_sys = convert(ODESystem, rxn_sys, combinatoric_ratelaws=false)
    prob = ODEProblem(ode_sys, u0, tspan, p)
    sol = solve(prob,Rosenbrock23())
    push!(vect, maximum(sol[2,:]))
end

plot(AHL, vect,
    xlabel = "AHL Concentration (uM)",
    ylabel = "GFP Concentration (uM)",
    label = "GFP", fmt = :png, title = "Dose Response Without Resource Conservation"
)
# savefig("no-resource-conservation-simulation")
```
